# Homework 3 – Survival Modeling & CLV Analysis
## Report

Our final survival model was the Log-Logistic AFT, which achieved the best balance of fit and interpretability (AIC = 1237.44, c-index ≈ 0.79). The significant predictors reveal a clear profile of churn risk. Higher churn risk is associated with subscribers who are unmarried, have no internet plan, and belong to less comprehensive service categories. Negative coefficients (time ratios < 1) shorten expected tenure: internet_Yes (0.43) and marital_Unmarried (0.64) indicate that subscribers in these groups churn substantially earlier. In contrast, positive coefficients increase survival time: E-service, Plus service, and Total service customers have time ratios between 2.01 and 2.54, meaning they remain subscribed for more than twice as long compared to the baseline. Age and address stability also slightly increase survival time, suggesting that older and more stable customers are more loyal.

Our CLV results highlight the most valuable segments. Retired customers (CLV ≈ 2146), Plus service users (CLV ≈ 1992), and married customers (CLV ≈ 1874) deliver the highest lifetime value. From a strategic standpoint, “valuable” subscribers are those with both high expected tenure and high predicted CLV, meaning that retention investments in these groups bring the strongest long-term returns. By contrast, internet users show lower CLV (≈ 1537), which is consistent with their significantly higher churn risk—this group should be prioritized for retention actions, such as proactive engagement, usage incentives, or bundled upgrades.

Assuming this dataset represents the population and using the survival curves, roughly 20–25% of subscribers exhibit strong downward survival probability over the next year, making them “at-risk.” With an average CLV around 1750, the potential value at risk is approximately 350–450 per customer annually. If we aim to intervene for ~100 at-risk customers, a reasonable annual retention budget would be $35,000–$45,000, depending on the intensity and cost of the interventions.

For retention, I would recommend: (1) targeted incentives for internet users and unmarried subscribers, (2) promoting upgrades to Plus/Total service bundles—which show substantially higher survival and CLV, (3) personalized outreach for customers early in their tenure when churn risk is steepest, and (4) creating senior-friendly or loyalty-oriented plans to retain high-value retired users.
